# Подключаем БД в работу: все три таблицы в базе данных создаем и джойним между собой и обратно возвращаем для работы в пандас

```python
con = sqlite3.connect('/content/sample_data/test.db')
df_abonent.to_sql('abonent', con,  index=False, if_exists='replace')
df_charge.to_sql('charge', con,  index=False, if_exists='replace')
df_traffic.to_sql('traffic', con,  index=False, if_exists='replace')
def select(sql):
  return pd.read_sql(sql, con)

sql = '''select a.Абонент as id, * from abonent a
join charge c on c.Абонент = a.Абонент
join traffic t on t.Абонент = a.Абонент
'''
df = select(sql)
df
```
