# Начинаем проведение Разведовательного Анализа Данных (EDA)

Оставляем только нужные столбцы

```python
columns = ['CustomerID', 'Gender', 'Location', 'Tenure_Months', 'Transaction_Date', 'Product_Category', 'Quantity', 'Avg_Price', 'Delivery_Charges', 'Coupon_Status', 'GST', 'Offline_Spend', 'Online_Spend', 'Discount_pct']
df = df[columns]
```

Пишем функцию подсчета профита, не учитывая доставку, т.к. она на заказчике:

```python
def use_coupon(row):
  if row['Coupon_Status'] == 'Used':
    discount = row['Quantity'] * row['Avg_Price'] * (row['Discount_pct'] / 100)
    TAX = (row['Avg_Price']) * row['GST']
    profit = row['Quantity'] * row['Avg_Price'] - TAX - discount
  else:
    TAX = (row['Quantity'] * row['Avg_Price']) * row['GST']
    profit = row['Quantity'] * row['Avg_Price'] - TAX
  return profit
```

```python
df['profit'] = df.apply(use_coupon, axis=1)
(df['profit'] < 0).value_counts()
```

Прибыль по локациям:

```python
(df.groupby('Location')['profit'].agg('sum')).sort_values(ascending=False)
```

Считаем финальный чек

```python
df['final_receipt'] = df['Quantity'] * df['Avg_Price']
```

Убеждаемся в корректности данных:

```python
x = df.groupby('CustomerID')['Tenure_Months'].agg(['min', 'max'])
(x['min'] == x['max']).value_counts()
```

Создаем новую таблицу подсчеты LTV для каждого заказчика

```python
LTV_table = df.groupby('CustomerID')[['Quantity', 'final_receipt', 'Tenure_Months']].mean()
LTV_table['LTV'] = LTV_table['Quantity'] * LTV_table['Tenure_Months'] * LTV_table['final_receipt']
LTV_table
```

Ищем корреляцию переменных исключая дискретные

# Проводим RFM-анализ

Разделяем чек на три группы по сумме: маленький, средний, большой.

```python
labels = ['small','medium','large']
df['final_receipt_bins'] = pd.cut(df['final_receipt'], bins = [0, 100, 300, float('infinity')], labels=labels)
```

```python
df['final_receipt_bins'].hist()
```

В данной диаграмме видны выбросы запредельных чеков по категориям товаров

```python
plt.scatter(df['final_receipt'], df['Product_Category'])
```

Разбиваем на три интервала "по времени сотрудничества с компанией":

"меньше года", "больше года, но меньше двух лет" и "больше двух лет"

```python
labels = ['<1', '>1 and <2', '>2']
df['Tenure_Months_bins'] = pd.cut(df['Tenure_Months'], bins=[0,12,24,float('infinity')], labels=labels)
df['Tenure_Months_bins'].hist()
```

Смотрим распределение значений по двум категориальным столбцам.

Большее количество заказчиков находятся в категории "Более двух лет".

Также 1599 из них в категории "Large" по среднему чеку.

```python
category_customers = df[['Tenure_Months_bins', 'final_receipt_bins']].value_counts().sort_index()
category_customers
```
